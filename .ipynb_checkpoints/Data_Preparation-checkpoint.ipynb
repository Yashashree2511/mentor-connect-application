```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("mentor_dataset_realistic_10000.xlsx")

df.head()
```

```python
# Remove extra spaces
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
# Convert to lowercase
df = df.apply(lambda x: x.str.lower() if x.dtype == "object" else x)

```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)

```

```python
# Clean skills column
df['skills'] = df['skills'].apply(
    lambda x: ','.join(set([i.strip() for i in str(x).split(',')])) if pd.notnull(x) else x
)
```

```python
# Handle missing values
df.dropna(subset=['domain', 'skills'], inplace=True)

df['bio'].fillna('unknown', inplace=True)
df['description'].fillna('unknown', inplace=True)
df['company_history'].fillna('unknown', inplace=True)
```

```python
# Remove junk characters
df.replace(r'[^a-zA-Z0-9, .+]', '', regex=True, inplace=True)
```

```python
# ---------------- FEATURE ENGINEERING ---------------- #

df['combined_text'] = (
    "domain: " + df['domain'].astype(str) + ". " +
    "skills: " + df['skills'].astype(str) + ". " +
    "experience: " + df['experience_years'].astype(str) + " years. " +
    "worked at: " + df['company_history'].astype(str) + ". " +
    "bio: " + df['bio'].astype(str) + ". " +
    "description: " + df['description'].astype(str)
)
```

```python
# ---------------- ENRICHMENT ---------------- #

def enrich(text):
    if "programming" in text:
        text += " software development coding backend apis"
    if "data science" in text:
        text += " machine learning ai analytics python"
    if "devops" in text:
        text += " ci cd docker kubernetes cloud"
    if "mobile development" in text:
        text += " android ios flutter react native"
    return text

df['combined_text'] = df['combined_text'].apply(enrich)
```

```python
df.to_csv("cleaned_mentor_data.csv", index=False)
```

```python
!pip install sentence-transformers
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
embeddings = model.encode(df['combined_text'].tolist(), show_progress_bar=True)
```

```python
def recommend_mentors(query, top_n=5):
    
    # Convert query to embedding
    query_embedding = model.encode([query])
    
    # Compute similarity
    similarity = cosine_similarity(query_embedding, embeddings)
    
    # Get top matches
    top_indices = np.argsort(similarity[0])[::-1][:top_n]
    
    # Fetch results
    results = df.iloc[top_indices][
        ['name', 'domain', 'skills', 'experience_years', 'company_history']
    ].copy()
    
    # Add similarity score
    results['similarity_score'] = similarity[0][top_indices]
    
    return results
```

```python
recommend_mentors("backend developer with java")
```

```python
recommend_mentors("machine learning engineer")
```

```python
import faiss
import numpy as np

# convert embeddings to numpy
embeddings_np = np.array(embeddings)

# normalize
embeddings_np = embeddings_np / np.linalg.norm(embeddings_np, axis=1, keepdims=True)

# create index
dimension = embeddings_np.shape[1]
index = faiss.IndexFlatIP(dimension)

# add data
index.add(embeddings_np)
```

```python
query = "machine learning mentor"

query_embedding = model.encode([query])

# convert to numpy
query_embedding = np.array(query_embedding)

# normalize
query_embedding = query_embedding / np.linalg.norm(query_embedding, axis=1, keepdims=True)

# search top 5
D, I = index.search(query_embedding, 5)
```

```python
import faiss

faiss.write_index(index, "faiss_index.bin")
```
